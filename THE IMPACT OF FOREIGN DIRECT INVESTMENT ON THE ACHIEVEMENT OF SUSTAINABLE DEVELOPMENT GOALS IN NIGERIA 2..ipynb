```python
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Correlation Analysis
correlation_matrix = df[['RGDP', 'FDI', 'INFL', 'BOP', 'OER']].corr()
print(correlation_matrix)
```

```python
# Plot Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Multiple Regression Analysis
# Define dependent and independent variables
X = df[['FDI', 'INFL', 'BOP', 'OER']]
y = df['RGDP']
```

```python
# Add constant to the model (intercept)
X = sm.add_constant(X)
```

```python
# Fit the model
model = sm.OLS(y, X).fit()
```

```python
# Print the summary of the regression
print(model.summary())
```

```python
# Define plotting function
def plot_regression(x, y, xlabel, ylabel, title):
    plt.figure(figsize=(10, 6))
    sns.regplot(x=x, y=y, scatter_kws={'s':50}, line_kws={'color':'red'})
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
# Plot regression for each independent variable against the dependent variable
plot_regression(df['FDI'], df['RGDP'], 'Foreign Direct Investment (FDI)', 'Real GDP (RGDP)', 'Regression of RGDP on FDI')
plot_regression(df['INFL'], df['RGDP'], 'Inflation Rate (INFL)', 'Real GDP (RGDP)', 'Regression of RGDP on INFL')
plot_regression(df['BOP'], df['RGDP'], 'Balance of Payment (BOP)', 'Real GDP (RGDP)', 'Regression of RGDP on BOP')
plot_regression(df['OER'], df['RGDP'], 'Official Exchange Rate (OER)', 'Real GDP (RGDP)', 'Regression of RGDP on OER')
```
