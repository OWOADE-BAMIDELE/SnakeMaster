
AN EXAMINATION OF THE INFLUENCE OF GLOBALIZATION ON NIGERIA FOREIGN POLICY (FIRSTBANK, LAGOS)


```python
import pandas as pd  
import matplotlib.pyplot as plt  

# Create a DataFrame for the survey data  
data = {  
    'Variable': [  
        'Age Group of the employees', 'Gender of the employees', 'Years of Employment of the employees', 'Current Position of the employees', 'Lagos branch of the employees'  
    ],  
    'Categories': [  
        ['Below 25', '26-35', '36-45', 'Above 45'],  
        ['Male', 'Female', 'Prefer not to say'],  
        ['Less than 1 year', '1-3 years', '4-6 years', 'More than 6 years'],  
        ['Executive', 'Management', 'Staff'],  
        ['FirstBank Marina', 'FirstBank Ikoyi', 'FirstBank Victoria Island',  
         'FirstBank Surulere', 'FirstBank Yaba', 'FirstBank Lekki']  
    ],  
    'Frequencies': [  
        [10, 70, 25, 7],  
        [40, 70, 2],  
        [15, 50, 25, 22],  
        [5, 45, 62],  
        [10, 12, 15, 8, 7, 6]  
    ]  
}  

df = pd.DataFrame(data)  

# Plot for each demographic characteristic  
for index, row in df.iterrows():  
    plt.figure(figsize=(8, 5))  
    plt.bar(row['Categories'], row['Frequencies'], color='skyblue')  
    plt.title(f'Distribution of {row["Variable"]}')  
    plt.xlabel(row["Variable"])  
    plt.ylabel('Frequency')  
    plt.xticks(rotation=45)  
    plt.tight_layout()  
    plt.show()
```

```python
import numpy as np  
import pandas as pd  
import seaborn as sns  
import matplotlib.pyplot as plt  

# Create a DataFrame for the correlation matrix  
data = {  
    'Globalization Impact on Foreign Policy (Y1)': [1, -0.45, 0.78, 0.55, 0.82],  
    'National Sovereignty (X1)': [-0.45, 1, -0.6, -0.35, -0.5],  
    'Foreign Economic Relations (X2)': [0.78, -0.6, 1, 0.48, 0.73],  
    'Technology Transfer (X3)': [0.55, -0.35, 0.48, 1, 0.68],  
    'Trade and Investment (X4)': [0.82, -0.5, 0.73, 0.68, 1]  
}  

# Create a DataFrame  
correlation_matrix = pd.DataFrame(data, index=[  
    'Globalization Impact on Foreign Policy (Y1)',  
    'National Sovereignty (X1)',  
    'Foreign Economic Relations (X2)',  
    'Technology Transfer (X3)',  
    'Trade and Investment (X4)'  
])  

# Set up the matplotlib figure  
plt.figure(figsize=(10, 8))  

# Draw the heatmap  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True, linewidths=.5)  

# Set title  
plt.title('Correlation Matrix between Globalization and Nigeria\'s Foreign Economic Relations', fontsize=16)  

# Show the plot  
plt.show()  
```

```python
import matplotlib.pyplot as plt  
import numpy as np  

# Data for the regression coefficients  
variables = ['Constant', 'National Sovereignty (X1)', 'Foreign Economic Relations (X2)',   
             'Technology Transfer (X3)', 'Trade and Investment (X4)']  
coefficients = [2.5, -0.45, 0.6, 0.25, 0.7]  
errors = [0.8, 0.12, 0.15, 0.1, 0.14]  

# Create a bar plot for the coefficients  
plt.figure(figsize=(10, 6))  
bars = plt.bar(variables, coefficients, yerr=errors, color=['blue', 'red', 'green', 'orange', 'purple'])  

# Add titles and labels  
plt.title('Multiple Regression Coefficients for Nigeria\'s Foreign Policy', fontsize=16)  
plt.xlabel('Variables', fontsize=14)  
plt.ylabel('Coefficient (B)', fontsize=14)  
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  

# Annotate the bars with their values  
for bar in bars:  
    yval = bar.get_height()  
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom' if yval > 0 else 'top')  

# Show the plot  
plt.tight_layout()  
plt.show()  
```
