THE EFFECTIVENESS OF SOCIAL MEDIA MARKETING IN ENHANCING CUSTOMER ENGAGEMENT IN SMALL AND MEDIUM ENTERPRISES (SMES) IN NIGERIA.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graph
categories = ['1-5', '6-10', '11-20', '21-50', '51-100', 'Over 100']
frequency_employees = [35, 25, 20, 10, 5, 5]

social_media = ['Facebook', 'Instagram', 'Twitter', 'LinkedIn', 'Other']
frequency_social_media = [40, 30, 15, 10, 5]
```

```python
# Plotting the number of employees
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(categories, frequency_employees, color='skyblue')
plt.title('Distribution of SMEs by Number of Employees')
plt.xlabel('Number of Employees')
plt.ylabel('Frequency')
```

```python

# Plotting the social media platforms actively used
plt.subplot(1, 2, 2)
plt.bar(social_media, frequency_social_media, color='salmon')
plt.title('Social Media Platforms Actively Used by SMEs')
plt.xlabel('Social Media Platform')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd  
from scipy import stats  

# Data from the table  
data = {  
    'Statement': [  
        "Our social media platforms engage customers",  
        "We receive more inquiries through social media",  
        "Social media helps us understand customer needs",  
        "We observe high engagement rates on social media",  
        "Social media has improved our customer relationships"  
    ],  
    'Mean': [3.55, 3.54, 3.57, 3.58, 3.52],  
    'Standard Deviation': [1.12, 1.10, 1.09, 1.11, 1.14],  
    'Sample Size': [100, 100, 100, 100, 100]  
}  

# Create a DataFrame  
df = pd.DataFrame(data)  

# Perform a one-sample t-test against a neutral score of 3  
neutral_score = 3  
t_statistic, p_value = stats.ttest_1samp(df['Mean'], neutral_score)  

# Output the results  
print("T-Statistic:", t_statistic)  
print("P-Value:", p_value)  

# Determine significance  
alpha = 0.05  
if p_value < alpha:  
    print("Reject the null hypothesis: Social media platforms significantly enhance customer engagement.")  
else:  
    print("Fail to reject the null hypothesis: No significant effect of social media on customer engagement.")  
```
