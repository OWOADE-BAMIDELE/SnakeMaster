```python
import pandas as pd  
import matplotlib.pyplot as plt  
```

```python
# Step 1: Data Preparation  
data = {  
    'Phage': ['phage 1', 'phage 2', 'phage 3', 'phage 4', 'phage 5'],  
    'Location': ['wvi IIJ'] * 5,  
    'Score Number': [2, 3, 2, 2, 2],  
    'E.coli 40': ['+', '+', '+', '+', '+'],  
    'E.coli 45': ['+', '+', '.', '-','-'],  
    'E.coli 498': ['-', '-', '-', '-', '-']  
}  

df = pd.DataFrame(data)  

# Step 2: Descriptive Statistics  
# Count the number of phages affecting each strain  
results = {  
    'Strain': ['E.coli 40', 'E.coli 45', 'E.coli 498'],  
    'Affected': [df['E.coli 40'].value_counts().get('+', 0),  
                 df['E.coli 45'].value_counts().get('+', 0),  
                 df['E.coli 498'].value_counts().get('+', 0)],  
    'Total': [len(df)] * 3  
}  

results_df = pd.DataFrame(results)  
results_df['Percentage'] = (results_df['Affected'] / results_df['Total']) * 100  

# Step 3: Visualization  
plt.figure(figsize=(10, 6))  
plt.bar(results_df['Strain'], results_df['Percentage'], color=['blue', 'orange', 'green'])  
plt.title('Percentage of Phages Affecting Each Strain')  
plt.xlabel('Strain')  
plt.ylabel('Percentage (%)')  
plt.ylim(0, 100)  
plt.grid(axis='y')  

# Show the percentage values on top of the bars  
for index, value in enumerate(results_df['Percentage']):  
    plt.text(index, value + 1, f"{value:.1f}%", ha='center')  

plt.show()  

# Display the results DataFrame  
print(results_df)  
```

```python
import pandas as pd  
import matplotlib.pyplot as plt  
import scipy.stats as stats  

# Step 1: Data Preparation  
# Table 1  
data1 = {  
    'Location': ['wvi IIJ'] * 5,  
    'Score Number': [2, 2, 2, 2, 2],  
    'E.coli 40': ['+', '+', '+', '+', '+'],  
    'E.coli 45': ['+', '+', '+', '+', '+'],  
    'E.coli 498': ['+', '+', '+', '+', '+']  
}  

# Table 2  
data2 = {  
    'Location': ['wvl IUZ'] * 5,  
    'Score Number': [2, 2, 2, 2, 2],  
    'E.coli 40': ['+', '+', '+', '+', '+'],  
    'E.coli 45': ['+', '+', '+', '+', '-'],  
    'E.coli 498': ['+', '+', '+', '+', '-']  
}  

# Table 3  
data3 = {  
    'Location': ['wvl 2AA'] * 5,  
    'Score Number': [3, 2, 2, 2, 2],  
    'E.coli 40': ['+', '+', '+', '+', '+'],  
    'E.coli 45': ['+', '+', '+', '+', '+'],  
    'E.coli 498': ['+', '-', '-', '-', '-']  
}  

# Combine all tables into a single DataFrame  
df1 = pd.DataFrame(data1)  
df2 = pd.DataFrame(data2)  
df3 = pd.DataFrame(data3)  
df = pd.concat([df1, df2, df3], ignore_index=True)  

```

```python
# View the dataframe
print(df) 
```

```python
# Step 2: Convert Effectiveness to Numeric  
df['E.coli 40'] = df['E.coli 40'].replace({'+': 1, '-': 0})  
df['E.coli 45'] = df['E.coli 45'].replace({'+': 1, '-': 0})  
df['E.coli 498'] = df['E.coli 498'].replace({'+': 1, '-': 0})  

# Step 3: Prepare Data for ANOVA  
anova_data = {  
    'Location': [],  
    'Strain': [],  
    'Score Number': []  
}  

# Collecting data for ANOVA  
for location in df['Location'].unique():  
    for strain in ['E.coli 40', 'E.coli 45', 'E.coli 498']:  
        scores = df[df['Location'] == location][strain].values  
        anova_data['Location'].extend([location] * len(scores))  
        anova_data['Strain'].extend([strain] * len(scores))  
        anova_data['Score Number'].extend(scores)  

# Create DataFrame for ANOVA  
anova_df = pd.DataFrame(anova_data)

# View the dataframe
print(df)
```

```python
# Step 4: Perform ANOVA  
anova_results = stats.f_oneway(  
    anova_df['Score Number'][anova_df['Strain'] == 'E.coli 40'],  
    anova_df['Score Number'][anova_df['Strain'] == 'E.coli 45'],  
    anova_df['Score Number'][anova_df['Strain'] == 'E.coli 498']  
)  

# View the ANOVA 
print(anova_results)

# Print ANOVA results  
print(f'ANOVA F-statistic: {anova_results.statistic}, p-value: {anova_results.pvalue}')  
```

```python
# Step 5: Visualize the Results  
plt.figure(figsize=(10, 6))  
plt.boxplot([  
    anova_df['Score Number'][anova_df['Strain'] == 'E.coli 40'],  
    anova_df['Score Number'][anova_df['Strain'] == 'E.coli 45'],  
    anova_df['Score Number'][anova_df['Strain'] == 'E.coli 498']  
], labels=['E.coli 40', 'E.coli 45', 'E.coli 498'])  

plt.title('Box Plot of Score Numbers by Strain')  
plt.ylabel('Score Number')  
plt.grid(axis='y')  
plt.show()  
```

```python
import pandas as pd  
import matplotlib.pyplot as plt  
import scipy.stats as stats  

# Step 1: Create DataFrames for each location  
data1 = {  
    'Phage': ['phage 1', 'phage 2', 'phage 3', 'phage 4', 'phage 5',   
              'phage 6', 'phage 7', 'phage 8', 'phage 9', 'phage 10'],  
    'Location 1 (cm)': [0.5, 0.65, 0.2, 0.7, 0.1, 0.15, 0.25, 0.55, 0.6, 0.62]  
}  

data2 = {  
    'Phage': ['phage 1', 'phage 2', 'phage 3', 'phage 4', 'phage 5',   
              'phage 6', 'phage 7', 'phage 8', 'phage 9', 'phage 10'],  
    'Location 2 (cm)': [0.25, 0.45, 0.55, 0.15, 0.6, 0.28, 0.33, 0.58, 0.63, 0.2]  
}  

data3 = {  
    'Phage': ['phage 1', 'phage 2', 'phage 3', 'phage 4', 'phage 5',   
              'phage 6', 'phage 7', 'phage 8', 'phage 9', 'phage 10'],  
    'Location 3 (cm)': [3.5, 0.55, 0.5, 0.6, 0.3, 0.22, 0.12, 0.75, 0.65, 0.62]  
}  

# Create DataFrames  
df1 = pd.DataFrame(data1)  
df2 = pd.DataFrame(data2)  
df3 = pd.DataFrame(data3)  

# Step 2: Perform T-tests  
# T-test between Location 1 and Location 2  
t_stat_1_2, p_value_1_2 = stats.ttest_ind(df1['Location 1 (cm)'], df2['Location 2 (cm)'])  

# T-test between Location 1 and Location 3  
t_stat_1_3, p_value_1_3 = stats.ttest_ind(df1['Location 1 (cm)'], df3['Location 3 (cm)'])  

# T-test between Location 2 and Location 3  
t_stat_2_3, p_value_2_3 = stats.ttest_ind(df2['Location 2 (cm)'], df3['Location 3 (cm)'])  

# T-test for all locations combined  
combined_data = pd.concat([df1['Location 1 (cm)'], df2['Location 2 (cm)'], df3['Location 3 (cm)']])  
t_stat_combined, p_value_combined = stats.ttest_1samp(combined_data, 0)  

# Step 3: Print T-test results  
print("T-test Results:")  
print(f"Location 1 vs Location 2: t-statistic = {t_stat_1_2}, p-value = {p_value_1_2}")  
print(f"Location 1 vs Location 3: t-statistic = {t_stat_1_3}, p-value = {p_value_1_3}")  
print(f"Location 2 vs Location 3: t-statistic = {t_stat_2_3}, p-value = {p_value_2_3}")  
print(f"Combined T-test: t-statistic = {t_stat_combined}, p-value = {p_value_combined}")  

# Step 4: Visualize the Results  
plt.figure(figsize=(12, 6))  
plt.boxplot([df1['Location 1 (cm)'], df2['Location 2 (cm)'], df3['Location 3 (cm)']],  
            labels=['Location 1', 'Location 2', 'Location 3'])  
plt.title('Box Plot of Measurements by Location')  
plt.ylabel('Measurement (cm)')  
plt.grid(axis='y')  
plt.show()  
```

```python
import pandas as pd  
import matplotlib.pyplot as plt  
import scipy.stats as stats  

# Step 1: Create DataFrames for each location  
data1 = {  
    'Phage': ['phage 1', 'phage 2', 'phage 3', 'phage 4', 'phage 5',   
              'phage 6', 'phage 7', 'phage 8', 'phage 9', 'phage 10'],  
    'Location 1 (cm)': [0.5, 0.65, 0.2, 0.7, 0.1, 0.15, 0.25, 0.55, 0.6, 0.62]  
}  

data2 = {  
    'Phage': ['phage 1', 'phage 2', 'phage 3', 'phage 4', 'phage 5',   
              'phage 6', 'phage 7', 'phage 8', 'phage 9', 'phage 10'],  
    'Location 2 (cm)': [0.25, 0.45, 0.55, 0.15, 0.6, 0.28, 0.33, 0.58, 0.63, 0.2]  
}  

data3 = {  
    'Phage': ['phage 1', 'phage 2', 'phage 3', 'phage 4', 'phage 5',   
              'phage 6', 'phage 7', 'phage 8', 'phage 9', 'phage 10'],  
    'Location 3 (cm)': [3.5, 0.55, 0.5, 0.6, 0.3, 0.22, 0.12, 0.75, 0.65, 0.62]  
}  

# Create DataFrames  
df1 = pd.DataFrame(data1)  
df2 = pd.DataFrame(data2)  
df3 = pd.DataFrame(data3)  

# Step 2: Prepare data for ANOVA  
anova_data = {  
    'Location': [],  
    'Measurement': []  
}  

# Collecting data for ANOVA  
for location, df in zip(['Location 1', 'Location 2', 'Location 3'], [df1, df2, df3]):  
    anova_data['Location'].extend([location] * len(df))  
    anova_data['Measurement'].extend(df.iloc[:, 1].values)  

# Create DataFrame for ANOVA  
anova_df = pd.DataFrame(anova_data)  

# Step 3: Perform ANOVA  
anova_results = stats.f_oneway(  
    anova_df['Measurement'][anova_df['Location'] == 'Location 1'],  
    anova_df['Measurement'][anova_df['Location'] == 'Location 2'],  
    anova_df['Measurement'][anova_df['Location'] == 'Location 3']  
)  

# Print ANOVA results  
print(f'ANOVA F-statistic: {anova_results.statistic}, p-value: {anova_results.pvalue}')  

# Step 4: Visualize the Results  
plt.figure(figsize=(12, 6))  
plt.boxplot([anova_df['Measurement'][anova_df['Location'] == 'Location 1'],  
             anova_df['Measurement'][anova_df['Location'] == 'Location 2'],  
             anova_df['Measurement'][anova_df['Location'] == 'Location 3']],  
            labels=['Location 1', 'Location 2', 'Location 3'])  
plt.title('Box Plot of Measurements by Location')  
plt.ylabel('Measurement (cm)')  
plt.grid(axis='y')  
plt.show()
```
